```python
from apiclient.discovery import build
```

```python
KEY = 'INSERT YOUR KEY HERE'
SERVICE = 'youtube'
VERSION = 'v3'
```

```python
yt_object = build(SERVICE,VERSION, developerKey = KEY)
```

```python
def yt_search(keyword, max_result = 50, order='relevance'):
    search_keyword = yt_object.search().list(q = keyword, part = "id, snippet", 
                                               maxResults = max_result).execute() 
    results = search_keyword.get("items", []) 
    vids = []
    for result in results:
        if result['id']['kind'] == 'youtube#video':
            vids.append(result)
    try:
        nextpagetok = search_keyword["nextPageToken"]
        return(nextpagetok, vids)
    except Exception as e:
        nextpagetok = "last_page"
        return(nextpagetok, vids)
    
        
        
    
    
```

```python
hist_dict = {'Video ID':[], 'Title':[],'Description':[], 'Category':[]}
```

```python
def VideoData(keyword,dicts,token = None):
    result = yt_search(keyword)
    vids = result[1]
    token = result[0]
    for vid in vids:
        dicts['Video ID'].append(vid['id']['videoId'])
        dicts['Title'].append(vid['snippet']['title'])
        dicts['Description'].append(vid['snippet']['description'])
        dicts['Category'].append(keyword)
    print("added: {}. New total {}".format(str(len(vids)),str(len(dicts['Video ID']))))
    return token
```

```python
counter = 0
tok = VideoData("history", dicts = hist_dict,token = None,)
while counter<38:
    tok = VideoData("history",dicts = hist_dict,token= tok)
    counter+=1
```
